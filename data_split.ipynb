```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('sepsis_final_RAW_continuous_13.csv')
```

### Train-test Split

```python
seed = 64
unique_trajs = df['traj'].unique()
train_traj, temp_traj = train_test_split(unique_trajs, test_size=0.4, random_state=seed)
val_traj, test_traj = train_test_split(temp_traj, test_size=0.5, random_state=seed)
train_df = df[df['traj'].isin(train_traj)]
val_df = df[df['traj'].isin(val_traj)]
test_df = df[df['traj'].isin(test_traj)]
train_indices = train_df.index
val_indices = val_df.index
test_indices = test_df.index
print(train_df.shape, val_df.shape, test_df.shape)
```

```python
os.makedirs(f'./{seed}/', exist_ok=True)
np.save('train_indices.npy', train_indices)
np.save('val_indices.npy', val_indices)
np.save('test_indices.npy', test_indices)
```

### Visualization Indices

```python
seed = 64
os.makedirs(f'./{seed}/', exist_ok=True)
train_indices = np.load('train_indices.npy')
test_indices = np.load('test_indices.npy')
train_outlier_labels = np.load('sepsis_final_RAW_continuous_13_train_outlier_labels_state_act.npy')
test_ourlier_labels = np.load('sepsis_final_RAW_continuous_13_test_outlier_labels_state_act.npy')
train_df = df.iloc[train_indices].reset_index(drop=True)
test_df = df.iloc[test_indices].reset_index(drop=True)
train_df_inlier = train_df.iloc[np.where(train_outlier_labels == 0)[0]]
test_df_inlier = test_df.iloc[np.where(test_ourlier_labels == 0)[0]]
train_df_inlier_ini = train_df_inlier[train_df_inlier['step'] == 0]
test_df_inlier_ini = test_df_inlier[test_df_inlier['step'] == 0]
train_df_inlier_indices = train_df_inlier.index
test_df_inlier_indices = test_df_inlier.index
```

```python
prefix = 'sepsis_final_RAW_continuous_13'
np.random.seed(seed)
train_selected_indices_ori = np.random.choice(train_df_inlier_indices, size=10000, replace=False)
train_selected_indices_ini = train_df_inlier_ini.index
test_selected_indices_ori = np.random.choice(test_df_inlier_indices, size=10000, replace=False)
test_selected_indices_ini = test_df_inlier_ini.index
np.save(f'{prefix}_train_indices_ori.npy', train_selected_indices_ori)
np.save(f'{prefix}_train_indices_ini.npy', train_selected_indices_ini)
np.save(f'{prefix}_test_indices_ori.npy', test_selected_indices_ori)
np.save(f'{prefix}_test_indices_ini.npy', test_selected_indices_ini)
```
